```python
from __future__ import absolute_import
from __future__ import division
from __future__ import print_function

import numpy as np
import pandas as pd
import tensorflow as tf
import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib inline
```

## Basic Training in TF

```python
x = tf.constant([[1], [2], [3], [4]], dtype=tf.float32)
y_true = tf.constant([[0], [-1], [-2], [-3]], dtype=tf.float32)

linear_model = tf.layers.Dense(units=1)

y_pred = linear_model(x)
loss = tf.losses.mean_squared_error(labels=y_true, predictions=y_pred)

optimizer = tf.train.GradientDescentOptimizer(0.01)
train = optimizer.minimize(loss)

init = tf.global_variables_initializer()

sess = tf.Session()
sess.run(init)
for i in range(10000):
  _, loss_value = sess.run((train, loss))
  if(i%100 == 0):
        print(loss_value)
```

```python
print(sess.run((y_pred,y_true)))
```

## Tensor and Variable Notes 

<ol><b>Tensors</b>
    <li>Tensors are the basic units of TF and they hold a value</li>
    <li>Tensors have ranks that implies their dimensionality</li>
    <li>Operations on tensors can be activated only by running sessions on them</li>
    <li>Different sessions will activate different values for a tensor if initialized as random or data</li>
</ol>

<ol><b>Variables</b>
    <li>Variables are container classes for tensors</li>
    <li>They also hold values but these mutations are visible across sessions</li>
    <li>Create a variable using get_variable("name",shape)</li>
    <li>Variables are usually initialised with glorot_uniform</li>
    <li>There are 2 collection lists - Global and Trainable</li>
    <li>Use global_variables_initializer() always before training to initialize all values</li>
    <li>Sharing variables by either passing it or by using scope</li>
    <li>Scope is useful when the same code needs to be used for different purposes like a cnn layer code but you want to create multiple layers with same config then use multiple scopes calling the same function!</li>
    <li>You can also reuse scope variables by saying scope.reuse_variables()</li>
</ol>
